```python
def verify_missing_fields(female_values, male_values, female_index, male_index):
    
    # Verify that all categories are equivalent for male and female
    fields_in_male_not_in_female = np.setdiff1d(male_index, female_index)
    female_index = list(female_index)
    female_values = list(female_values)
    for i in fields_in_male_not_in_female:
        female_index.append(i)
        female_values.append(0)

    fields_in_female_not_in_male = np.setdiff1d(female_index, male_index)
    male_index = list(male_index)
    male_values = list(male_values)
    for i in fields_in_female_not_in_male:
        male_index.append(i)
        male_values.append(0)
        
    return female_values, male_values, female_index, male_index
```

```python
def statistic_about_women(male_values, female_values):
    
    percentage_fem = [(female_values[i]/(female_values[i]+val))*100 for i, val in enumerate(male_values)]
    # print('percentage_fem: ', percentage_fem)
    
    return percentage_fem
```
